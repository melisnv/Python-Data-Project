## K Means Cluster Creation

It is time to create the Cluster labels!

**Importing KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

**Creating an instance of a K Means model with 2 clusters.**

```python
kmeans = KMeans(n_clusters=2)
```

**Fitting the model to all the data except for the Private label.**

```python
df = df.drop('Unnamed: 0',axis=1)
dataf = df.drop('Private',axis =1)
```

```python
dataf
```

```python
kmeans.fit(dataf)
```

**What are the cluster center vectors?**

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

## Evaluation

There is no perfect way to evaluate clustering if I don't have the labels, however since this is just an exercise, I do have the labels, so I take advantage of this to evaluate my clusters, I'm aware that usually won't have this luxury in the real world.

**Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
def converter(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

**Creating a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(df['Cluster'],kmeans.labels_))
```
